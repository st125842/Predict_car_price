```python
def remove_mileage(x):
        
    if isinstance(x, str) :
        value = x.split(' ')[0]
        if value != '':
            return float(value)        
def get_brand(x):
    return x.split(' ')[0]
```

```python
df['mileage'] = df['mileage'].apply(remove_mileage)
df['engine'] = df['engine'].apply(remove_mileage)
df['max_power'] = df['max_power'].apply(remove_mileage)
df['name'] = df['name'].apply(get_brand)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
y_train = np.log(y_train)
```

```python
print(X_train['max_power'].mean())
print(X_train['engine'].median())
print(X_train['seats'].mode())
```

```python
X_train['max_power'].fillna(X_train['max_power'].mean(),inplace=True)
X_train['engine'].fillna(X_train['engine'].median(),inplace=True)
X_train['seats'].fillna(5,inplace=True)
```

```python
col = X_train.columns
```

```python
X_test['max_power'].fillna(X_train['max_power'].mean(),inplace=True)
X_test['engine'].fillna(X_train['engine'].median(),inplace=True)
X_test['seats'].fillna(5,inplace=True)
```

```python
from sklearn.preprocessing import StandardScaler

# feature scaling helps improve reach convergence faster
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test  = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression  #we are using regression models
from sklearn.svm import SVR
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor
from xgboost import XGBRegressor
# Libraries for model evaluation

# models that we will be using, put them in a list
algorithms = [LinearRegression(), SVR(), GradientBoostingRegressor(random_state = 0), DecisionTreeRegressor(random_state = 0), 
              RandomForestRegressor(n_estimators = 100, random_state = 0),XGBRegressor(random_state=0)]

# The names of the models
algorithm_names = ["Linear Regression", "SVR", "GradientBoostingRegressor", "Decision-Tree Regressor", "Random-Forest Regressor","xgboost"]
```

```python
from sklearn.model_selection import KFold, cross_val_score

#lists for keeping mse
train_mse = []
test_mse = []

#defining splits
kfold = KFold(n_splits=5, shuffle=True)

for i, model in enumerate(algorithms):
    scores = cross_val_score(model, X_train, y_train, cv=kfold, scoring='r2')
    print(f"{algorithm_names[i]} - Score: {scores}; Mean: {scores.mean()}")
```

```python
from sklearn.model_selection import KFold, cross_val_score

#lists for keeping mse
train_mse = []
test_mse = []

#defining splits
kfold = KFold(n_splits=5, shuffle=True)

for i, model in enumerate(algorithms):
    scores = cross_val_score(model, X_train, y_train, cv=kfold, scoring='neg_mean_squared_error')
    print(f"{algorithm_names[i]} - Score: {scores}; Mean: {scores.mean()}")
```

```python
model = XGBRegressor()
```

```python
model.fit(X_train,y_train)
```

### Inference

```python
pred = model.predict(X_test)
```

```python
pred = np.exp(pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
print("MSE: ", mean_squared_error(y_test, pred))
print("r2: ", r2_score(y_test, pred))
```

```python
rmse = np.sqrt(mean_squared_error(y_test, pred))
print(f"RMSE: {rmse:.4f}")
```

## Which algorithm performs well? Which does not? Why?

I got higher accurasy with XGBoost Model. And Low accuracy is linear regression.

Linear regression is a basic model that works well when the data follows a straight, clear pattern.
But in real life, data is often messy and complicated, so linear regression can't always give good results.

XGBoost, uses many decision trees to model data more flexibly.It also uses weights and a learning rate to control how each tree contributes, allowing it to adapt better to the structure of real-world data.
